Question 5:

```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
for name, grade in marks.items():
    print(f'Grade for {name} is {grade}')
    plt.bar(name, grade)
    plt.title(f'Bar Chart of {name} Grade')
    plt.show()
```

```python
y1 = np.array([3, 5, 11])
y2 = np.array([2, 17, 5])
plt.subplot(2, 3, 1)
plt.title("Plot w/ Multiple Lines")
plt.plot(y1)
plt.plot(y2)

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 2)
plt.title("Bar Chart")
plt.bar(x,y)

y = np.array([35, 25, 25, 15])

plt.subplot(2, 3, 3)
plt.title("Pie Chart")
plt.pie(y)

x = np.array([20, 25, 30, 35, 40, 45, 50, 55, 60, 65])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])


plt.subplot(2, 3, 4)
plt.title("Plot w/ Grids")
plt.plot(x, y)
plt.grid()

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.subplot(2, 3, 5)
plt.title("Scatter Plot")
plt.scatter(x, y)

x = np.random.normal(100, 10, 150)


plt.subplot(2, 3, 6)
plt.title("Histogram")
plt.hist(x)

plt.show()
```
